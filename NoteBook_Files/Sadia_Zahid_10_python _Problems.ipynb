# Problem 1

```python
def calculate_grade(avg_percentage):
    if avg_percentage >= 90:
        return "A"
    elif avg_percentage >= 80:
        return "B"
    elif avg_percentage >= 70:
        return "C"
    elif avg_percentage >= 60:
        return "D"
    else:
        return "F"

sub1 = int(input("Enter marks for subject 1: "))
sub2 = int(input("Enter marks for subject 2: "))
sub3 = int(input("Enter marks for subject 3: "))
sub4 = int(input("Enter marks for subject 4: "))
sub5 = int(input("Enter marks for subject 5: "))

total_sum = sub1 + sub2 + sub3 + sub4 + sub5
avg_percentage = total_sum / 5

grade = calculate_grade(avg_percentage)

print("Total Sum:", total_sum)
print("Average Percentage:", avg_percentage)
print("Grade:", grade)
```

# Problem 2

```python
def analyze_number(num):
    if num > 0:
        nature = "Positive"
    elif num < 0:
        nature = "Negative"
    else:
        nature = "Zero"

    if num == 0:
        parity = "Neither even nor odd"
    elif num % 2 == 0:
        parity = "Even"
    else:
        parity = "Odd"
        
    return nature, parity

user_input = int(input("Enter an integer: "))

nature, parity = analyze_number(user_input)

print(f"{nature}, {parity}")
```

# Problem 3

```python
def add(a, b):
    return a + b
def subtract(a, b): 
    return a - b
def multiply(a, b): 
    return a * b
def divide(a, b):
    return a / b if b != 0 else "Error: Division by zero"

def calculator():
    print("Select operation: +, -, *, /")
    choice = input("Enter choice: ")
    
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    if choice == '+':
        print("Result:", add(num1, num2))
    elif choice == '-':
        print("Result:", subtract(num1, num2))
    elif choice == '*':
        print("Result:", multiply(num1, num2))
    elif choice == '/':
        print("Result:", divide(num1, num2))
    else:
        print("Invalid input")

calculator()
```

```python
def get_stats(numbers):
    numbers.sort()
    
    largest = numbers[-1]
    smallest = numbers[0]
    total_sum = sum(numbers)
    average = total_sum / len(numbers)
    second_largest = numbers[-2]
    
    return largest, smallest, total_sum, average, second_largest

user_input = input("Enter numbers separated by spaces: ")
num_list = [float(i) for i in user_input.split()]

if len(num_list) < 2:
    print("Please enter at least two numbers.")
else:
    largest, smallest, total_sum, average, second_largest = get_stats(num_list)

    print("Largest:", largest)
    print("Smallest:", smallest)
    print("Sum:", total_sum)
    print("Average:", average)
    print("Second Largest:", second_largest)
```

# Problem 6

```python
def count_words(sentence):
    words = sentence.split()
    frequency = {}
    
    for word in words:
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
            
    return frequency

text = input("Enter a sentence: ")
result = count_words(text)

print(result)
```

# Problem 7

```python
def validate_password(password):
    if len(password) < 8:
        return "Invalid: Must be at least 8 characters long."
    
    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_special = any(not char.isalnum() for char in password)
    
    if not (has_upper and has_lower):
        return "Invalid: Must contain both uppercase and lowercase letters."
    if not has_digit:
        return "Invalid: Must contain at least one digit."
    if not has_special:
        return "Invalid: Must contain at least one special character."
    
    return "Valid"

user_pwd = input("Enter your password: ")
print(validate_password(user_pwd))
```

# Problem 8

```python
prices = {"apple": 1.0, "banana": 0.5, "milk": 2.5, "bread": 2.0}
cart = {}

def show_menu():
    print("\n--- Shopping Cart ---")
    print("1. Add Item | 2. Remove Item | 3. Update Quantity | 4. View Cart & Total | 5. Exit")

def view_cart():
    total = 0
    print("\nYour Cart:")
    for item, quantity in cart.items():
        price = prices.get(item, 0)
        subtotal = price * quantity
        total += subtotal
        print(f"{item}: {quantity} x ${price} = ${subtotal:.2f}")
    print(f"Total Bill: ${total:.2f}")

while True:
    show_menu()
    choice = input("Select an option: ")
    
    if choice == '1':
        item = input("Enter item name: ").lower()
        if item in prices:
            qty = int(input("Enter quantity: "))
            cart[item] = cart.get(item, 0) + qty
        else:
            print("Item not in store.")
            
    elif choice == '2':
        item = input("Enter item to remove: ").lower()
        if item in cart:
            del cart[item]
        else:
            print("Item not in cart.")
            
    elif choice == '3':
        item = input("Enter item to update: ").lower()
        if item in cart:
            cart[item] = int(input("Enter new quantity: "))
        else:
            print("Item not in cart.")
            
    elif choice == '4':
        view_cart()
        
    elif choice == '5':
        break
    else:
        print("Invalid choice.")
```

# Problem 9

```python
def divide_numbers():
    try:
        num1 = float(input("Enter the numerator: "))
        num2 = float(input("Enter the denominator: "))
        
        result = num1 / num2
        print(f"The result is: {result}")
        
    except ValueError:
        print("Error: Invalid input. Please enter numeric values only.")
        
    except ZeroDivisionError:
        print("Error: You cannot divide by zero.")
        
    finally:
        print("Operation attempted.")

divide_numbers()
```
