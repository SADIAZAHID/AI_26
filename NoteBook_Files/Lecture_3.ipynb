## Positinal Argumets

```python
# function with Arguments 

def greet( name):
    print("Hello",name)
```

```python
# function Call

greet(name)
```

```python
# function definition 

def Age(age):
    print(" I am ",age,"years old")
```

```python
Age(80)
```

```python
def greet(name,age):
    print("hello",name,"  I am ",age,"Years old")
    
```

```python
greet("sadia",60)
```

## Keywords Arguments

```python
greet(age=70,name="sara")
```

## Default Arguments

```python
def Age(name,age=23):
    print(name," ,I am ",age,"years old")
```

```python
Age("sadia")
```

```python
Age("sadia",88)
```

### Variable Length Arguments

```python
def vla(*arg):
    print(arg)
```

```python
vla(1,2,3)
```

```python
def add(*args):
    print(sum(args))
```

```python
add(1,2,3,4,5,6)
```
