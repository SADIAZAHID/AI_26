## Overview
Lambda functions are small, anonymous functions defined with the `lambda` keyword. They are useful for simple operations and functional programming.

## Key Characteristics
- **Anonymous**: No function name
- **Single expression**: Limited to one expression
- **Inline**: Defined where used
- **Functional**: Often used with higher-order functions

```python
# Basic lambda functions
square = lambda x: x**2
add = lambda x, y: x + y
is_even = lambda x: x % 2 == 0

print(f"Square of 5: {square(5)}")
print(f"Sum of 3 and 7: {add(3, 7)}")
print(f"Is 6 even? {is_even(6)}")

# Lambda with built-in functions
numbers = [1, 2, 3, 4, 5, 6]
squares = list(map(lambda x: x**2, numbers))
evens = list(filter(lambda x: x % 2 == 0, numbers))

print(f"Squares: {squares}")
print(f"Even numbers: {evens}")
```

## Summary
Lambda functions are useful for:
- Simple operations
- Functional programming
- Callback functions
- Sorting and filtering
