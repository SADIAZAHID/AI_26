## Polymorphism

```python
class Cat:
    def speaks(self):
        print("Meow")

class Dog:
    
    def speak(self):
        print("Barks")

    def animal_sound(animal):
        animal.speak()

    animal_sound(Cat())
    animal_sound(Dog())
```

```python
class Base:
    def execute(self):
        print("Base: Operation started")

class Logger(Base):
    def execute(self):
        print("Logger: Saving to logs")
        super().execute()

class Authenticator(Base):
    def execute(self):
        print("Authenticator: Checking permissions")
        super().execute()

class Processor(Logger, Authenticator):
    def execute(self):
        print("Processor: Starting work")
        super().execute()
```

```python
obj = Processor()
obj.execute()
```

## MultiLevel Inheritance 

```python
# Level 1: The Foundation Team
class Foundation:
    def build_foundation(self):
        print("1. Foundation is laid.")

# Level 2: The Structure Team (Inherits from Foundation)
class Structure(Foundation):
    def build_walls(self):
        self.build_foundation() # Uses the foundation
        print("2. Walls are built.")

# Level 3: The Finishing Team (Inherits from Structure)
class House(Structure):
    def paint(self):
        self.build_walls() # Uses the structure
        print("3. House is painted.")
```

```python
my_house = House()
my_house.paint()
```

## Herarchical Inheritance

```python
class PaymentProcessor:
    def base_process(self):
        print("1. Connecting to secure bank server.")
        print("2. Validating transaction details.")

class CreditCardPayment(PaymentProcessor):
    def pay(self):
        self.base_process()
        print("3. Processing via Credit Card gateway (Visa/Mastercard).")

class PayPalPayment(PaymentProcessor):
    def pay(self):
        self.base_process()
        print("3. Processing via PayPal API.")
```

```python
# Industry Usage:
cc_pay = CreditCardPayment()
cc_pay.pay()
```

## Hybrid Inheritance

```python
# 1. Base Class (Common for all computers)
class Computer:
    def power_on(self):
        print("System booting up...")

# 2. Hierarchical: Two different types of computers
class Laptop(Computer):
    def open_lid(self):
        print("Lid opened.")

class Desktop(Computer):
    def start_fan(self):
        print("Fan spinning.")

# 3. Multiple Inheritance: Adding specific features (mixins)
class WiFiModule:
    def connect_wifi(self):
        print("WiFi connected.")

class BluetoothModule:
    def connect_bluetooth(self):
        print("Bluetooth connected.")

# 4. The Hybrid Class (Combined)
class Ultrabook(Laptop, WiFiModule, BluetoothModule):
    def start_work(self):
        self.power_on()
        self.open_lid()
        self.connect_wifi()
        print("Ready for work!")
```

```python
# Execution
my_laptop = Ultrabook()
my_laptop.start_work()
```

## Types of Method In class(oop)

```python
class student:
    
    #class variable
    company="corvit"
    
    #constructor method
    def __init__(self,name):
        self.name=name

    #instance method 
    def change_name(self,name):
        self.name=name
        
    #class method
    @classmethod
    def change_company(cls,name):
        cls.company=name

    #static method 
    @staticmethod
    def fee(total,bonus_percent):
        fees=total-(bonus_percent*total/100)
        print(fees)
        return fees

    
```

```python
std=student("ALi")
std1=student("Ahmed")
```
