## Unit Converter App

## main functionality

```python
def meter(m):
    feet = 3.28*m
    inch = 39.27*m
    print(f"{m} meters={feet} ft and {inch} inches")

def feet(f):
    meter=0.308*f
    inch=12*f
    print(f"{f} feets={meter} metre and {inch} inches")

def inch(i):
    meter=0.0254*i
    feet=0.083*i
    print(f"{i} inches={meter} ft and {feet} inches")
```

```python
meter(1)
```

```python
feet(1)
inch(1)
```

```python
while(True):
    unit=input("Welcome to unit Converter\n Choose \n m for meter \n f for feet\n i for inch \n q for Quit")
    if (unit=="q" or unit=='q'):
        break

    val=float(input("Enter Value to convert"))
    
    if(unit=="M" or unit =='m'):
        meter(val)
        
    elif((unit=="F" or unit =='f')):
         feet(val)
        
    elif((unit=="I" or unit =='i')):
         inch(val)
        
    else:
        print("Please enter valid character m,f,i or q")
    
```

```python
from tkinter import *
from tkinter import messagebox

def convert(unit_type):
    try:
        # Get the value from the entry field
        val = float(entry.get())
        
        # Perform calculation
        if unit_type == "m":
            result = f"{val} meters is {val * 3.28084:.2f} feet"
        elif unit_type == "f":
            result = f"{val} feet is {val * 12:.2f} inches"
        elif unit_type == "i":
            result = f"{val} inches is {val * 2.54:.2f} cm"
        
        messagebox.showinfo("Result", result)
    except ValueError:
        messagebox.showerror("Error", "Please enter a valid number")

# Setup the main window
root = Tk()
root.title("Unit Converter")
root.geometry("300x250")

Label(root, text="Enter Value:").pack(pady=5)
entry = Entry(root)
entry.pack(pady=5)

# Buttons for conversion
Button(root, text="Meter to Feet", command=lambda: convert("m")).pack(fill='x', padx=20)
Button(root, text="Feet to Inch", command=lambda: convert("f")).pack(fill='x', padx=20)
Button(root, text="Inch to Cm", command=lambda: convert("i")).pack(fill='x', padx=20)

# Quit button
Button(root, text="Quit", command=root.destroy).pack(pady=20)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def start_app():
    root = tk.Tk()
    root.title("Unit Converter")
    
    # Entry
    entry = tk.Entry(root)
    entry.pack()

    # Function inside
    def run_conv(unit):
        try:
            val = float(entry.get())
            if unit == 'm': res = val * 3.28
            messagebox.showinfo("Result", f"Result: {res}")
        except:
            messagebox.showerror("Error", "Valid number enter karein")

    # Buttons
    tk.Button(root, text="Convert Meter", command=lambda: run_conv('m')).pack()
    tk.Button(root, text="Quit", command=root.destroy).pack()
    
    root.mainloop()


start_app()
```

```python
start_app()
```

# Password Validator

```python
def pass_check(key):
    if len(key)<8:
        return "invalid"
    up =False
    lo= False
    di= False
    sp= False
    for char in key :
        if char.isupper():
            up =True
        if char.islower():
            lo= True
        if char.isdigit():
            di=True
        if char in ("!@#$%&()_+*~"):
            sp=True
        if (up and lo and di and sp):
            return "Valid"
        return "Invalid"
     
```

```python
pass_check("334")
```
