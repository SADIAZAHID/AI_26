### 1. What is the difference between a variable and a keyword in Python? 

#### Variable: 

A variable is a name used to store data.  

Its value can be changed during program execution.  

Example:

name="sadia ",
age= 70

#### Keyword:

A keyword is a reserved word in Python with a predefined meaning.
Keywords cannot be used as variable names.
Examples:
if, else, for, while, def, return, class.
