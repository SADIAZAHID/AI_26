## Find the index of 2 numbers from sum whose sum is equal to target

```python
nums=[8,2,4,7,5]
target =9
if nums[0]+nums[1]==target:
   print("(0,1)")
elif nums[0]+nums[2]==target:
    print("(0,2)")
elif nums[0]+nums[3]==target:
   print("(0,3)")
elif nums[0]+nums[3]==target:
   print("(0,4)")
elif nums[1]+nums[2]==target:
   print("(1,2)")
elif nums[1]+nums[3]==target:
   print("(1,3)")
elif nums[1]+nums[4]==target:
   print("(1,4)")
elif nums[2]+nums[3]==target:
   print("(2,3)")
elif nums[2]+nums[4]==target:
   print("(2,4)")
elif nums[3]+nums[3]==target:
   print("(3,3)")
elif nums[3]+nums[4]==target:
   print("(3,4)")
else:
    print("nothing")
```
