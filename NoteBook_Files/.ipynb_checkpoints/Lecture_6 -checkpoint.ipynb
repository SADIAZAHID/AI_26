## File Handling

```python
f=open("Files/file.txt",'r') #for read a file and if there is no file it gives error
```

```python
f=open("Files/file.txt",'w') #for write a file 
```

```python
f.write("hello from Code")# this a way to write text in a file 
f.close()
```

```python
f=open("Files/file.txt",'r')
print(f.read())
f.close()
```
